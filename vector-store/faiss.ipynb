```python
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
from langchain_community.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=0
)

text = """
Artificial Intelligence (AI) is a branch of computer science that focuses on 
creating machines and software capable of performing tasks that normally require 
human intelligence, such as learning, reasoning, problem-solving, decision-making, 
and understanding natural language. AI systems use technologies like machine learning, 
deep learning, and neural networks to analyze large amounts of data, recognize patterns, 
and make accurate predictions. Today, AI is widely used in various industries, including 
healthcare, finance, education, transportation, and entertainment. Applications such as virtual assistants, 
recommendation systems, self-driving cars, fraud detection, and medical diagnosis have significantly improved 
efficiency and productivity. As AI continues to evolve, it is transforming the way people live and work by automating 
repetitive tasks, enhancing decision-making, and enabling innovative solutions to complex problems. Despite its many 
benefits, AI also raises important ethical concerns related to privacy, bias, transparency, and job displacement,
 making responsible development and deployment essential for ensuring that its impact remains beneficial to society."""

docs = splitter.split_text(text)
print(docs[0])
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
%pip install faiss-cpu
```

```python
vector_store = FAISS.from_texts(
    texts=docs,
    embedding=embeddings
)
```

```python
vector_store.similarity_search_with_score(
    query='What is Artificial Intelligence?',
    k=2
)
```
