```python
from langchain_groq import ChatGroq
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
import requests

from dotenv import load_dotenv
load_dotenv()
```

```python
# tool creation

@tool 
def multiply(a : int , b:int) -> int:
    """multiply two numbers"""
    return a*b
```

```python
print(multiply.invoke({'a':3 , 'b' : 9}))
```

```python
# tool binding 
model = ChatGroq(model="llama-3.3-70b-versatile")
```

```python
model_with_tool = model.bind_tools([multiply])
```

```python
model_with_tool.invoke("hiii how are you")
```

```python
query = HumanMessage("can you please multiply 2 and 3 ")
```

```python
messages = [query]
```

```python
results= model_with_tool.invoke(messages)
```

```python
messages.append(results)
```

```python
messages
```

```python
#tool execution
tool_result = multiply.invoke(results.tool_calls[0])
```

```python
messages.append(tool_result)
```

```python
model_with_tool.invoke(messages).content
```
