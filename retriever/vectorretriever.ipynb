```python
from langchain_community.vectorstores import Chroma 
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.documents import Document
```

```python
documents = [
    Document(page_content='Langchain helps developers build langchain applications easily'),
    Document(page_content='chroma is a vector database optimised for llm-based search'),
    Document(page_content='embedding converts text into high dimensional vectors'),
    Document(page_content='Groq provides powerful embeddings model')

]
```

```python
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

vector_store = Chroma.from_documents(
    documents=documents,
    embedding= embedding_model,
    collection_name= 'my_collection'
)
```

```python
retriever = vector_store.as_retriever(search_kwargs={'k' :2})
```

```python
query = 'what is chroma used for?'

docs = retriever.invoke(query)
```

```python
for i , doc in enumerate(docs):
    print(f"\n----Result {i+1}-----")
    print(f"--content : {doc.page_content}....")
```

```python
res  = vector_store.similarity_search(
    query='what is chroma used for?',
    k=2
)

for i , doc in enumerate(res):
    print(f"\n----Result {i+1}-----")
    print(f"--content : {doc.page_content}....")

```
