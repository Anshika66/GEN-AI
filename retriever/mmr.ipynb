```python
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.documents import Document
```

```python
documents = [
    Document(page_content='Langchain helps developers build langchain applications easily'),
    Document(page_content='chroma is a vector database optimised for llm-based search'),
    Document(page_content='embedding converts text into high dimensional vectors'),
    Document(page_content='Groq provides powerful embeddings model')

]
```

```python
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
vector_store = FAISS.from_documents(
    documents=documents,
    embedding=embedding_model
)

```

```python
results = vector_store.as_retriever(
    search_type="mmr",    #this enables mmr 
    search_kwargs={'k':3 , 'lambda_mult':0.5}

)
```

```python
query='what are chroma is used for ?'
docs = results.invoke(query)


for i , doc in enumerate(docs):
    print(f"\n----Result {i+1}-----")
    print(f"--content : {doc.page_content}....")
```
